```python
# import libraries
import pandas as pd
import numpy as np
import os

# create path variable
path = r'C:\Users\justi\Downloads\05-2024 Instacart Basket Analysis_JB\05-2024 Instacart Basket Analysis'

# use path and retrieve prepared "Orders" data
df_ords = pd.read_csv(os.path.join(path, '02. Data', 'Prepared Data', 'orders_checked.csv'), index_col = False)

# use path and retrieve prepared "Products" data
df_prods = pd.read_csv(os.path.join(path, '02. Data', 'Prepared Data', 'products_checked.csv'), index_col = False)

# use path and retrieve "orders_products_prior" csv
df_ords_prior = pd.read_csv(os.path.join(path, '02. Data', 'Original Data', 'orders_products_prior.csv'), index_col = False)
```

```python
# 'df_ords' and 'df_ords_prior' dataframes share the same column called 'order_id'. Use .merge() and 'indicator = True' functions to create a new dataframe called 'df_ords_merged'
df_ords_merged = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)
df_ords_merged
```
