## 1b. Create "busiest_day" column.

```python
# Update 'order_dow' column name--> 'order_day_of_week' using .rename({}), inplace = true functions
df.rename(columns = {'order_dow' : 'order_day_of_week'}, inplace = True)
df
```

```python
# Use .value_counts(dropna = False) to find frequency of how many times each 'order_day_of_week' is recorded
# Looks like users order the most on Day 0 (busiest day) and order the lease on Day 4 (least busy)
df['order_day_of_week'].value_counts(dropna = False)
```

```python
#  For-loop function will run through every row in 'order_day_of_week' column, compare its value with what are busiest and slowest days, and assign it the corresponding string value.

result = []
for value in df["order_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Use 'result' to to create new column to call it 'busiest day' in the dataframe to view output in context.
df['busiest_day'] = result
df
```

# 2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

## 2a. Create For-Loop function of new request

```python
# Update for-loop function to make Days 0-1 = 'busiest days', 3-4 = 'slowest days', and the rest = 'regularly busy'

result_2 = []

for value in df["order_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

```python
# Now replace 'result' for-loop with newest 'result_2' for-loop
df['busiest_day'] = result_2
```

```python
# Rename 'busiest_day' column to 'busiest_days' using .rename({}), inplace=True functions
df.rename(columns={'busiest_day': 'busiest_days'}, inplace=True)
```

# 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Check frequency of which hour is ranked highest in "order_hour_of_day" column.

df['order_hour_of_day'].value_counts(dropna = False)
```

```python
# "Most orders" = Top 8 hours = 14, 10, 13, 11, 12, 15, 16, 9 
# "Average orders" = Middle 8 hours = 17, 8, 18, 19, 20, 7, 21, 22
# "Fewest orders" = Bottom 8 hours = 23, 6, 0, 1, 5, 4, 3, 2
# Create for-loop if statement labeling each specified hour as “Most orders,” “Average orders,” or “Fewest orders.”

result_3 = []

for value in df["order_hour_of_day"]:
  if value in [14, 10, 13, 11, 12, 15, 16, 9]:
    result_3.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 4, 3, 2]:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
result_3
```

```python
# Create new column "busiest_period_of_day" in the dataframe.
df['busiest_period_of_day'] = result_3
df
```
