## Q5. Determine the meaning behind a value of 4 in the "department_id" column within the df_prods dataframe using a data dictionary.

```python
df_prods.head()
```

```python
# Transposing departments dataframe ('df_dep') using '.T' function and labeling it as a new variable: 'df_dep_t'
df_dep_t = df_dep.T
df_dep_t
```

```python
# Using '.reset_index()' function after 'df_dep_t' to create an index to this specific dataframe
df_dep_t.reset_index()
```

```python
# Using '.iloc[0]' function to locate the first line item from 'df_dep_t' dataframe and label this as 'new_header' variable
new_header = df_dep_t.iloc[0]
new_header
```

```python
# 'df_dep_t' dataframe starts with row 0. What does '[1:]' mean? This is telling Python to copy everything from the first row onward from 'df_dep_t' dataframe into a new dataframe: 'df_dep_t_new'. The 1 represents row 1 in the dataframe.
df_dep_t_new = df_dep_t[1:]
df_dep_t_new
```

```python
# 2 new variables were created in the previous steps: 'new header' and 'df_dep_t_new'. Now marry the 2 variables using '.column' function to assign names to the columns in a dataframe. TLDR: “Take the values stored in the new_header variable and put them in the df_dep_t_new dataframe as the names of its columns.” 
df_dep_t_new.columns = new_header
df_dep_t_new
```

```python
# Using '.to_dict()' function to transform 'df_dep_t_new' dataframe into dictionary format and saved it in a new variable: 'data_dict'. The argument: ('index'), tells Python to use the numbered rows as the key values for the entries in the dictionary.
data_dict = df_dep_t_new.to_dict('index')
data_dict
```

```python
print("Answer to Q5. 4= Produce. Needed to transpose departments data, fix its headers, and make a data dictionary to cross-reference with below dataframe ('df_prods')'s 'department_id' column.")
```

## Q6. The sales team in your client’s organization wants to know more about breakfast item sales. Create a subset containing only the required information.

```python
# Above data dictionary shows the department breakfast = 14. Enclosing df_prods['department_id']==14 inside df_prods[] tells Python to map the true/false procedure  onto the columns within the df_prods dataframe. TLDR: Provide a list of only those values within df_prods that are Department 14.
df_breakfast = df_prods[df_prods['department_id'] == 14]
df_breakfast
```

```python
avg_breakfast_price = df_breakfast['prices'].mean()

print("Answer to Q6: The average price for breakfast products is:", avg_breakfast_price)
print("This calculation comes from above subset of 1116 breakfast items from the 'Products dataset'.")
```
