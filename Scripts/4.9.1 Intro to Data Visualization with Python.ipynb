# 4. Wrangle the data so that it follows consistent logic; for example, rename columns with illogical names and drop columns that don’t add anything to your analysis.

```python
# Column titles are different formats. Rename the columns to make it all lowercase and uniform using .rename(columns={:}) function
customers.rename(columns = {'First Name' : 'first_name'}, inplace = True)
customers.rename(columns = {'Surnam' : 'last_name'}, inplace = True)
customers.rename(columns = {'Gender' : 'gender'}, inplace = True)
customers.rename(columns = {'Age' : 'age'}, inplace = True)
customers.rename(columns = {'STATE' : 'state'}, inplace = True)
customers.rename(columns = {'fam_status' : 'marital_status'}, inplace = True)
```

# 5. Complete the fundamental data quality and consistency checks you’ve learned throughout this Achievement; for example, check for and address missing values and duplicates, and convert any mixed-type data.

```python
# Check each column's data types
customers.dtypes
```

```python
# Checking for mixed data-types in each column
for col in customers.columns.tolist():
  weird = (customers[[col]].map(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

```python
# 'first_name' is mixed-type data. Convert to string using .astype('str') function
customers['first_name'] = customers['first_name'].astype('str')
```

```python
# 'user_id' is int64, while it does not apepar as mixed-type data, this is a numerical substitute to represent the user. Convert to string using .astype('str') function
customers['user_id'] = customers['user_id'].astype('str')
```

```python
# Validate that 'first_name' and 'user_id' in the dataframe are now "object"
customers.dtypes
```

```python
# Validate that 'first_name' does not appear as result for below mixed-type data check code
for col in customers.columns.tolist():
  weird = (customers[[col]].map(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

```python
# Check for missing data in 'customers' dataframe using .isnull().sum() functions
customers.isnull().sum()
```

```python
# Check for duplicate data in 'customers' dataframe using .duplicated() function by calling it 'customers_dups'
customers_dups = customers[customers.duplicated()]
```
