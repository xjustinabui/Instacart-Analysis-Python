# 4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Create new column "max_order" using .groupby([]) .transform(np.max) on the columns 'user_id' and 'order_number'
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking for new column 'max_order' appears in far right of dataframe
df_ords_prods_merge.head(20)
```

```python
# Assign loyalty flag using .loc[]
# If user orders above 40 = 'Loyal customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# If user orders between 10-40 = 'Regular customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# If user orders 10 or less = 'New customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check loyalty flags using .value_counts(dropna = False) function
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check Loyalty flag in dataframe of first 10 results
df_ords_prods_merge[['user_id', 'loyalty_flag','max_order']].head(10)
```

# 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

## If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
## If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Creating new column called 'average spend' using .groupby and .transform(np.mean) functions
df_ords_prods_merge['average_spend'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Creating flag 'low spender' for average spend < 10
df_ords_prods_merge.loc[df_ords_prods_merge['average_spend'] <10,'spender_flag'] ='Low spender'
```

```python
# Creating flag 'high spender' for average spend >= 10
df_ords_prods_merge.loc[df_ords_prods_merge['average_spend'] >= 10, 'spender_flag'] = 'High spender'
```

```python
# Check frequency count of spender flags using .value_counts(dropna = False) function
df_ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

```python
# View first 10 results of dataframe with 'spender_flag' column data now
df_ords_prods_merge[['user_id', 'prices', 'average_spend', 'spender_flag']].head(10)
```

# 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

## If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
## If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
## If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Create a column for frequent users called 'Customer_frequency' using 'days_since_prior_order' column, .groupby([]) and .transform(np.median) functions
df_ords_prods_merge['Customer_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Create 'frequency flag' on for 'days_since_prior_order' > 20 = 'Non-frequent customer' 
df_ords_prods_merge.loc[df_ords_prods_merge['Customer_frequency'] > 20, 'frequency_flag'] = 'Non_frequent_customer'
```

```python
# Create 'frequency flag' on for 'days_since_prior_order' is between 10-20 = 'Regular customer' 
df_ords_prods_merge.loc[(df_ords_prods_merge['Customer_frequency'] >10) & (df_ords_prods_merge['Customer_frequency'] <=20),'frequency_flag'] = 'Regular customer'
```

```python
# Create 'frequency flag' on for 'days_since_prior_order' is 10 days or less = 'Frequent customer' 
df_ords_prods_merge.loc[df_ords_prods_merge['Customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
# Check frequency flags using .value_counts(dropna = False)
df_ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# View first 10 results of dataframe with 'customer_frequency' and 'frequency_flag' columns data now
df_ords_prods_merge[['user_id','days_since_prior_order', 'Customer_frequency', 'frequency_flag']].head(10)
```
